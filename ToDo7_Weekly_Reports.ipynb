## **0 - Preparação do relatório**

```python
#importando a  biblioteca Pandas
import pandas as pd

#importando o arquivo csv que está no direito com gastos diários da loja
#a coluna 'Dia' será o índex explícito do DataFrame
loja_df = pd.read_csv("/content/drive/MyDrive/RESILIA/To Do/Todo_7_Relatórios_Semanais/Todo7_relatorio .csv", index_col='Dia', decimal=',')
loja_df
```

```python
# Acrescentando a coluna com os ganhos diários:

loja_df['Receita Diária'] = [2200, 2420.50, 3391, 5322, 4898.50, 4200, 3893]
loja_df

```

```python
# Acrescentando os impostos diários totais (alíquota de 7% ao dia):

loja_df['Impostos'] = [(0.07*2200), (0.07*2420.50), (0.07*3391), (0.07*5322), (0.07*4898.50), (0.07*4200), (0.07*3893)]
loja_df
```

## **1. Análise dos ganhos(receitas)**

**Observações:**

* O **maior ganho** diário obtido foi registrado na **quinta-feira**  com o total de R$5.322,00, o que representa 20,2 % do ganho semanal.

* Por outro lado, **o menor ganho** diário ocorreu na **segunda-feira**, com o total de R$2.200,00 representando 8,4% do ganho semanal.

* Na maioria dos dias da semana, entre **quinta-feira a domingo**, o ganho diário **superou** a **receita média semanal de R$ 3.760,71**.

---

**Sugestões:**

* Recomenda-se analisar a origem dos ganhos para **identificar quais os itens/categorias que mais venderam** e portanto, sustentaram a maior parte da receita da loja durante a semana. Essa informação também será útil para análise de estoque.

* Recomenda-se também **identificar os itens com baixa saída**. Em seguida, seria indicado realizar pesquisas de satisfação com os clientes para entender suas preferências e necessidades a fim de aumentar a venda de tais produtos futuramente.

* Buscar **identificar possíveis razões para o aumento de ganhos na quinta-feira**: houve alguma campanha ou promoção? Assim como identificar a composição das vendas nesse dia e vendedores responsáveis.

```python
# Estatísticas básicas da tabela inicial:
loja_df.describe()
```

```python
#Cálculo da receita total semanal:
receita_total = loja_df['Receita Diária'].sum(axis = 0)
print( 'A receita total semanal é de R$',receita_total)
```

```python
#Cálculo da receita diária em percentual (%) sobre a receita total semanal (ordem descrescente):

receita_em_percentual = loja_df['Receita Diária']/receita_total*100
receita_em_percentual.sort_values(ascending=False)
```

```python
#Cálculo da receita média semanal:
receita_media = loja_df['Receita Diária'].mean(axis =0)
print ('O ganho médio semanal é de R$', receita_media)
```

## **2. Análise das despesas**

**Observações:**

* **Quarta-feira** foi o dia da semana em que houve mais gastos (3.184,37)
o que representa **32,6% dos gastos semanais**(R$ 9.747,39).

* A maior parte dos gastos da quarta-feira são representados pela **categoria 'Outros'**, com o total de R$ 2.310,00.

* Aproximadamente **50% dos gastos** da semana foram efetuados em apenas dois dias: **quarta e quinta-feira**.

* Igualmente, **a média de gastos de R$ 1.392**, foi **superada** apenas na **quarta e quinta e feira**.

* Em relação as categorias, a denominada **"Outros"** é de fato a que mais contribui negativamente para as despesas da loja, **representando 37% dos gastos semanais**. Em segundo lugar temos  a categoria "Comida" representando 23,8% dos gastos.

---



**Sugestões:**

* Recomenda-se fortemente **verificar os itens que compõe a categoria 'Outros**' e entender porque há tanta **variação** nessa categoria e  as **razões de ter sido tão expressiva na quarta-feira.** A seguir é necessário ver meios para diminuir ou eliminar gastos desnecessários;

* Recomenda-se **verificar também as variações com gastos em comida** (Há diferente quantidade de funcionários na loja em alguns dias? As refeiçoes estão sendo compradas em lugares diferentes? Ou apenas os preços que subiram?).

* Há também **variação nos gastos com transporte** (15% dos gastos semanais), é também interessante buscar entender porque isso acontece e se há maneiras de economizar nesse quesito.



```python
# A despesa diária total:
loja_df['Despesa Diária Total'] = loja_df.loc[:'Domingo',['Limpeza','Comida','Transporte', 'Outros','Impostos']].sum(axis = 1)
loja_df
```

```python
#A despesa semanal total:

despesa_total = loja_df['Despesa Diária Total'].sum(axis = 0)
print('A despesa total dessa semana foi de R$', despesa_total)
```

```python
# A despesa diária total em percentual (ordem decrescente):
despesa_em_percentual = loja_df['Despesa Diária Total']/despesa_total*100
despesa_em_percentual.sort_values(ascending=False)

```

```python
# A média semanal de todas as despesas:

despesa_media = loja_df['Despesa Diária Total'].mean(axis = 0)
print( 'A média de gastos dessa semana foi de R$', despesa_media)
```

```python
# Soma total das despesas por categoria (ordem decrescente):
despesas_por_categoria = loja_df[['Limpeza','Comida','Transporte','Outros','Impostos']].sum()
despesas_por_categoria.sort_values(ascending=False)
```

```python
# Soma total das despesas por categoria em percentual (ordem decrescente):
despesas_categoria_percentual = despesas_por_categoria/despesa_total*100
despesas_categoria_percentual.sort_values(ascending=False)
```

## **3. Análise dos lucros**

**Observações:**

* A **receita líquida** diária e semanal foram **positivas**. A receita líquida representa os ganhos após deduções como os impostos, neste caso.
* O **lucro líquido** diário e semanal foram **positivos**. O lucro líquido representa o montante após subtrairmos as despesas totais dos ganhos obtidos.
* A **média de lucro** diário foi de R$ 2.368,23.
* Em geral, ao longo da semana, o lucro diário apresentou **crescimento**. Porém, apresentou flutuações:
  - Se manteve em crescimento até quarta-feira quando teve uma queda abrupta;
  - Voltou a crescer entre a quinta e sexta-feira (dia mais lucrativo);
  - Porém, apresentou queda no final de semana.
* As despesas tiveram grande impacto na lucratividade da quarta-feira (ficando bem aquém da média: 206,63).


---
**Recomendações:**

* A loja apresentou rentabilidade expressiva durante essa semana, porém é necessário **avaliar o fluxo de caixa** e entender qual a previsão de receber os ganhos com as vendas (a prazo ou à vista);

* Recomenda-se **continuar efetuando analises semanais** para avaliar **tendências** nos quesistos analisados: ganhos, despesas e lucros.


```python
# Receita Líquida Diária = receita bruta das vendas subtraindo-se deduções, neste caso apenas os impostos

loja_df['Receita Diária Sem Impostos'] = loja_df['Receita Diária']-loja_df['Impostos']
loja_df

```

```python
# Receita Líquida Semanal: receita total semanal sem os impostos:
receita_sem_impostos = loja_df['Receita Diária Total Sem Impostos'].sum(axis =0)
print(' A receita total semanal retirados os impostos é de R$',receita_sem_impostos)
```

```python
#Lucro líquido médio:

lucro_liquido.mean()
```

```python
# Qual o dia mais lucrativo:
lucro_liquido.sort_values(ascending=False)
```
